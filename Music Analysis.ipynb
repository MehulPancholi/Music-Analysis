```python
from pydub import AudioSegment

song = AudioSegment.from_mp3("DaniCalifornia.mp3")
```

# Cool Functions

```python
raw = song.raw_data #Raw audio data in bytes
len(raw)
```

```python
channel_count = song.channels # (1 means mono, 2 means stereo)
channel_count
```

```python
duration = song.duration_seconds # length of the song
duration
```

```python
frames_per_second = song.frame_rate # Common values are 44100 (CD), 48000 (DVD), 22050, 24000, 12000 and 11025
frames_per_second
```

```python
backwards = song.reverse()
backwards.export("DaniCaliforniaBackwards.mp3", format="mp3")
```

```python
# pydub does things in milliseconds
ten_seconds = 10 * 1000
first_10_seconds = song[:ten_seconds]
last_5_seconds = song[-5000:]
```

```python
song[:ten_seconds]
```

```python
quieter = song - 15 # Quiet the song by 15 decibals
quieter[:ten_seconds]
```

```python
# Merge the first ten seonds and the last 5 seconds
merged = first_10_seconds.append(last_5_seconds)
merged
```

```python
# Merge the first ten seonds and the last 5 seconds with a crossfade
mergedCrossfade = first_10_seconds.append(last_5_seconds, crossfade=5000)
mergedCrossfade
```

```python
# Overlap first 10 seconds with last 5 
playedTogther = first_10_seconds.overlay(last_5_seconds, times=2)
playedTogther
```

```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt

# read audio samples
input_data = read("DaniCalifornia.wav")
audio = input_data[1]
# plot the first 1000240 samples
plt.plot(audio[0:1000240])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```
